```python
import cv2
import time
from playsound import playsound
import supervision as sv
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
tracker = sv.ByteTrack()

SEUIL_TEMPS = 5  
suivi_durees = {}  

def jouer_bip():
    playsound("C:/Windows/Media/notify.wav") 

# Capturer la webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Erreur : Impossible d'ouvrir la webcam.")
    exit()

# Boucle principale de d√©tection et suivi en temps r√©el
while True:
    ret, frame = cap.read()
    if not ret:
        print("Erreur : Impossible de capturer la vid√©o.")
        break

    # D√©tection d'objets avec YOLO
    results = model(frame)

    # Extraction des r√©sultats sous forme de tableaux numpy
    boxes = results[0].boxes.xyxy.cpu().numpy()
    confidences = results[0].boxes.conf.cpu().numpy()
    class_ids = results[0].boxes.cls.cpu().numpy()

    # Cr√©ation de l'objet Detections pour ByteTrack
    detections = sv.Detections(
        xyxy=boxes, 
        confidence=confidences, 
        class_id=class_ids
    )

    tracked_objects = tracker.update_with_detections(detections)

    for track_id, (x1, y1, x2, y2), cls_id in zip(tracked_objects.tracker_id, tracked_objects.xyxy, tracked_objects.class_id):
        x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])

        label = f"ID {track_id}: {model.names[int(cls_id)]}"
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # D√©clencher l'alarme si l'objet reste d√©tect√© plus de SEUIL_TEMPS secondes
        if track_id not in suivi_durees:
            suivi_durees[track_id] = time.time()
        else:
            duree = time.time() - suivi_durees[track_id]
            if duree >= SEUIL_TEMPS:
                cv2.putText(frame, "ALERTE !", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)
                jouer_bip()

    # Afficher le flux vid√©o
    cv2.imshow("D√©tection et Suivi avec YOLO et Alarme", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
